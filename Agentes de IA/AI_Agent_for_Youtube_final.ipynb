## Acessando o Youtube

E agora vamos mostrar o quão fácil é de acessar vídeos de youtube utilizando LangChain!

```python
from langchain_community.document_loaders import YoutubeLoader

url = 'https://www.youtube.com/watch?v=Q2DjTJP3tQI'

loader = YoutubeLoader.from_youtube_url(url, language = ['pt'])
documentos = loader.load()

documento = ''
for doc in documentos:
  documento = documento + doc.page_content

documento
```

```python
template = ChatPromptTemplate.from_messages([
   ('system', 'Você é um assistente amigável que possui as seguintes informações para formular uma resposta: {informações}'),
   ('user', '{input}')
])

chain_youtube = template | chat

pergunta = 'Quando seria interessante fazer long & short segundo autor'

resposta = chain_youtube.invoke({'informações': documento, 'input': pergunta})

print(resposta.content)
```
